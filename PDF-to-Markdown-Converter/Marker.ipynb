```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered
from pathlib import Path
import os

print("Marker 라이브러리 로드 완료!")
```

```python
import tkinter as tk
from tkinter import filedialog

# ---------------------------------------------------
# [실행 부분] 윈도우 파일 탐색기를 엽니다.
# ---------------------------------------------------

print("파일 탐색기를 여는 중입니다... (창이 뒤에 숨어있을 수 있습니다)")

# 1. 못생긴 tkinter 기본 창 숨기기
root = tk.Tk()
root.withdraw()
root.wm_attributes('-topmost', 1) # 창을 맨 앞으로 가져오기

# 2. 파일 선택 창 띄우기

pdf_file = filedialog.askopenfilename(
    title="변환할 PDF 파일을 선택하세요",
    filetypes=[("PDF Files", "*.pdf")]
)
# 3. 변환 실행
if pdf_file:
    print(f"선택된 파일: {pdf_file}")
    text, images = convert_pdf_to_markdown(pdf_file)
else:
    print("파일 선택이 취소되었습니다")
```

```python
def convert_pdf_to_markdown(pdf_path, output_dir="output"):
    """
    PDF 파일을 마크다운으로 변환하는 함수
    """
    # 출력 디렉토리 생성 (오타 수정: makedirss -> makedirs)
    os.makedirs(output_dir, exist_ok=True)

    print(f"PDF 변환 시작 : {pdf_path}")

    # 파일 존재 여부 확인
    if not os.path.exists(pdf_path):
        print(f"오류: '{pdf_path}' 파일을 찾을 수 없습니다.")
        return None, None

    print("모델을 로딩하는 중... (최초 실행 시 시간이 걸릴 수 있습니다)")
    try:
        # Converter 생성
        converter = PdfConverter(
            artifact_dict = create_model_dict(),
        )
        # Pdf 변환 실행
        rendered = converter(pdf_path)

        # 텍스트와 이미지 추출
        text, _, images = text_from_rendered(rendered)

        # 마크다운 파일 저장
        output_filename = Path(pdf_path).stem + ".md"
        output_path = os.path.join(output_dir, output_filename)

        # (문법 수정: . -> ,)
        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(text)

        print(f"\변환 완료!")
        print(f"출력 파일: {output_path}")
        print(f"텍스트 길이: {len(text):,} 글자")
        print(f"이미지 개수: {len(images)}")


        # 미리보기 출력
        print("\n" + "="*60)
        print("변환된 텍스트 미리보기:")
        print("=" * 60)
        print(text[:500] + "...\n")
        return text, images

    except Exception as e:
        print(f"오류 발생: {str(e)}")
        # (오타 수정: none - > None)
        return None, None
```

```python
if os.path.exists(pdf_file):
    text, images = convert_pdf_to_markdown(pdf_file)
else:
    print(f"파일을 찾을 수 없습니다: {pdf_file}")
```
