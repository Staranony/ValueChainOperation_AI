```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered
from pathlib import Path
import os

print("Marker   !")
```

```python
def convert_pdf_to_markdown(pdf_path, output_dir="output"):
    """
    PDF     
    
    Args:
        pdf_path: PDF  
        output_dir :  
    """
    #   
    os.makedirss(output_dir, exist_ok=True)

    print(f"PDF   : {pdf_path}")
    print("  ... (      )")
    try:
        # Converter 
        converter = PdfConverter(
            artifact_dict = create_model_dict(),
        )

        # Pdf  
        rendered = converter(pdf_path)

        #   
        text, _, images = text_from_rendered(rendered)

        #   
        output_filename = Path(pdf_path).stem + ".md"
        output_path = os.path.join(output_dir, output_filename)

        with open(output_path, 'w'. encoding='utf-8') as f:
            f.write(text)

        print(f"\n  !")
        print(f" : {output_path}")
        print(f" : {len(text):,} ")
        print(f" : {len(images)}")

        #   ( 500)

        print("\n" + "="*60)
        print("  :")
        print("="*60)
        print(text[:500] + "...\n")
        
        return text, images
    
    except Exception as e:
        print(f" : {str(e)}")
        return None, none
    
# 4.  

pdf_file = "your_file.pdf"

#  
text, images = convert_pdf_to_markdonw(pdf_file)


# 5.  PDF   

def convert_multiple_pdfs(pdf_folder, output_folder="output")
    """
       PDF   
    
    Args:
        pdf_folder: PDF    
        output_folder :    
    
    """
    os.makedirs(output_folder, exist_ok=True)

    # PDF   
    pdf_files = list(Path(pdf_folder).glob("*.pdf"))

    if not pdf_files:
        print(f"{pdf_folder}  .")
        return
    print(f"{len(pdf_files)} PDF   \n")

    converter = PdfConverter(
        artifact_dict=create_model_dict(),
    )
    for i, pdf_path in enumerate(pdf_files, 1):
        print(f"[{i}/{len(pdf_files)}]   : {pdf_path.name}")

        try:
            rendered = converter(str(pdf_path))
            text, _, images = text_from_rendered(rendered)

            output_filename = pdf_path.stem + ".md"
            output_path = os.path.join(output_folder, output_filename)

            with open(output_path, 'w', encoding='utf-8') as f:
                f.write(text)

            print(f"")


```
